```python
# import libraries and reading our CSV

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create

# Read in the data

df=pd.read_csv(r'C:\Users\Damian\Documents\Datasets\Pokemon2.csv')
```

```python
#Doing a quick read on the head of the dataset to get a basic idea of what it looks like
#Also extending the max number of columns seen so we can scroll to see all

pd.options.display.max_columns=50
df.head()
```

```python
# Data types for columns
# Sometimes it's very useful to have a reference on what all of the data types in the dataset, so we are doing a quick check here

df.dtypes
```

```python
# Checking for any missing data to see which columns have missing data and what percent

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# We will now take the data we've looked at to ask a few questions
# Is there a correlation between how much a pokemon weighs and its HP?
# Scatter plot w/ weight vs hp

plt.scatter(x=df['weight_kg'], y=df['hp'])

plt.title('Weight vs HP')
plt.xlabel('Weight(kg)')
plt.ylabel('HP')

plt.show()
```

```python
# Since we checked out weight, let's see if height affects HP.
# Scatter plot w/ height vs hp using seaborn


sns.regplot(x='height_m', y='hp', data=df, scatter_kws={"color":"red"}, line_kws={"color":"green"})
```

```python
# So it does appear there may be a slight correlation between height and hp, but not much.
```

```python
# Is there a correlation between how much a pokemon weighs and its speed?
# Scatter plot w/ weight vs speed

plt.scatter(x=df['weight_kg'], y=df['speed'])

plt.title('Weight vs Speed')

plt.xlabel('Weight(kg)')

plt.ylabel('Speed')

plt.show()
```

```python
# Similar to the previous "Weight vs HP" chart, the Weight vs Speed yielded similar results with no definitive direction
```

```python
#Getting a basic understanding of the correlation of all the columns

df.corr()
```

```python
#Visualizing the correlation w/ a heatmap

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=False)

plt.title('Correlation Matrix')

plt.xlabel('Pokemon Features')

plt.ylabel('Pokemon Features')

plt.show()
```

```python
#Since we have the correlation of all of our data points, let's focus on just the stat attributes

df[['attack','sp_attack','hp','speed','defense','sp_defense']].corr()
```

```python
#Let's now visualize the data above to get a better understanding of it

stat_corr = df[['attack','sp_attack','hp','speed','defense','sp_defense']].corr()

sns.heatmap(stat_corr, 
            annot=True, 
            linewidths=1,
            cmap="Reds",
            cbar_kws={"orientation": "horizontal"})

#names the graph along with x + y axis
plt.title('Correlation Matrix')
plt.xlabel('Pokemon Stats')
plt.ylabel('Pokemon Stats')

plt.show()
```

```python


sns.heatmap(
    df[df['secondary_type']!='None'].groupby(['primary_type', 'secondary_type']).size().unstack(),
    linewidths=1,
    annot=True,
    cmap="Reds",
    cbar_kws={"orientation": "horizontal"})

plt.xticks(rotation=90)
plt.show()
```

```python
#Lets find the amount of pokemon types and visualize them in a pie chart

fig = plt.figure(figsize=(20,20))

fig.add_subplot(221)
df['primary_type'].value_counts().plot(kind='pie', 
                                       autopct='%1.1f%%',
                                       pctdistance=0.9)

fig.add_subplot(223)
df['secondary_type'].value_counts().plot(kind='pie', 
                                       autopct='%1.1f%%',
                                       pctdistance=0.9)

plt.show()
```

```python
#Let's see those results in a different chart with a number attached to them now
#First chart for the primary type
sns.catplot(
    y='primary_type',
    data=df,
    kind='count',
    order=df['primary_type'].value_counts().index,
    height=4,
    aspect=2,
    color='red'
    ).set_axis_labels('Number of Pokemon', 'Primary Type')

#Plotting the secondary type
sns.catplot(
    y='secondary_type',
    data=df,
    kind='count',
    order=df['secondary_type'].value_counts().index,
    height=4,
    aspect=2,
    color='blue'
    ).set_axis_labels('Number of Pokemon', 'Secondary Type')
```

```python
#Assigning colors to the chart and counting legendary/non legendary
colors = ["green", "red"]
legendary = df[df['is_legendary']==True]
nLegendary = df[df['is_legendary']==False]

#charting the data and counting the english name to establish a count then labeling and formatting chart
numLeg = [legendary['english_name'].count(),nLegendary['english_name'].count()]
pieChart = plt.pie(numLeg,
                 labels= ['Legendary', 'Not Legendary'], 
                 autopct ='%1.1f%%', 
                 shadow = True,
                 colors=colors,          
                 explode=(0, 0.3))
```
