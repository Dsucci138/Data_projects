```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create

# Read in the data

df=pd.read_csv(r'C:\Users\Damian\Documents\Datasets\movies.csv')
```

```python
#Look at data head

df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for columns

df.dtypes
```

```python
# changing data type for columns

df = df.fillna(0)

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
# Grabbing the correct year from "released" and making new column 

df['correctyear'] = df['released'].astype(str).str.split(',').str[1]
df['correctyear'] = df['correctyear'].astype(str).str[:5]
df.head()
```

```python
# Sorting columns by gross

df = df.sort_values(by=['gross'], inplace = False, ascending=False)
```

```python
#Allows for df to display all rows/columns + scroll to see all
pd.set_option('display.max_rows', None)
```

```python
# Dropping duplicates in company column
# To drop dupes of whole DF- df.drop_duplicates()

df_company = df['company'].astype(str).drop_duplicates().sort_values(ascending=False)

df_company.head(10)
```

```python
# Scatter plot w/ budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Film Budget')

plt.show()
```

```python
df.head()
```

```python
# plot budget vs gross w/ seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"green"})
```

```python
#Shows correlation -  pearson by default
#Other correlation methods-Pearson, Kendall, spearman

df.corr()
```

```python
#Visualizing the correlation w/ a heatmap

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df.head()
```

```python
#Visualizing the numerized correlation w/ a heatmap

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
#Stacks correlation chart

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
#Stacks correlation chart in pairs

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
#Finds correlation pairs over .5

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
